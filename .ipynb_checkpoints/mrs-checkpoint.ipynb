```python
import numpy as np
import pandas as pd 

```

```python
pk = pd.read_csv("pokemon.csv")
```

```python
pk.reset_index()
```

```python
df = pk[pk['Name'].str.contains('Darkrai')]
print(df.id)
```

```python
 pk = df.append(pk)
```

```python
pk = pk.drop_duplicates(subset='id')
```

```python
pk['Type_1'].value_counts()
```

```python
stat = pd.DataFrame(pk[['''HP','Attack','Defense','Sp. Atk','Sp. Def','Speed']])
```

```python
stat
```

```python
def cal_distance(x):      
    a = x.to_numpy()
    b = stat.to_numpy()    
    distance = np.sqrt(sum([(a - b) ** 2 for a, b in zip(a, b[0])]))
    return distance
```

```python
pk['distance'] = pk[['HP','Attack','Defense','Sp. Atk','Sp. Def','Speed']].apply(cal_distance,axis=1)
```

```python
pk = pk.sort_values('distance',ascending=True)
```

```python
pk.reset_index()
```
